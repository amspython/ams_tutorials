```python
%matplotlib inline

import numpy as np

a = (np.arange(10)+np.random.random(10))/2
b = (-1*np.arange(10)-np.random.random(10))/2
c = np.random.random(100).reshape(10,10)*10
```

```python
plt.figure()
plt.plot(a)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(a)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax1.plot(a)
ax2 = fig.add_subplot(2,1,2)
ax2.plot(b)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.plot(a)
ax2 = fig.add_subplot(1,2,2)
ax2.plot(b)
```

add_subplot(#rows, #cols, position)

```python
fig = plt.figure()
ax1 = fig.add_subplot(3,1,1)
ax1.plot(a)
ax2 = fig.add_subplot(3,1,2)
ax2.plot(b)
ax3 = fig.add_subplot(3,1,3)
im = ax3.matshow(c)
fig.colorbar(im, ax=ax3)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot(a)
ax2 = fig.add_subplot(2,2,2)
ax2.plot(b)
ax3 = fig.add_subplot(2,2,4)
im = ax3.matshow(c)
fig.colorbar(im, ax=ax3)
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure()
#create a grid of 2 rows, 2 columns for plots
gs = gridspec.GridSpec(2,2)
#put a plot in the first row, first column
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(a)
```

```python
fig = plt.figure()
#create a grid of 2 rows, 2 columns for plots
gs = gridspec.GridSpec(2,2)
#put a plot in the first row, first column
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(a)
#put a plot in the second row, first col
ax2 = fig.add_subplot(gs[1,0])
ax2.plot(b)
```

```python
fig = plt.figure()
#create a grid of 2 rows, 2 columns for plots
gs = gridspec.GridSpec(2,2)
#put a plot in the first row, first column
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(a)
ax1.set_aspect('equal')
#put a plot in the second row, first col
ax2 = fig.add_subplot(gs[1,0])
ax2.plot(b)
ax2.set_aspect('equal')
#have it second col, all rows
ax3 = fig.add_subplot(gs[:,1])
im = ax3.imshow(c)
fig.colorbar(im, ax=ax3, fraction=.045)
```

```python
fig = plt.figure()
#create a grid of 2 rows, 2 columns for plots
gs = gridspec.GridSpec(5,5)
#put a plot in the first row, first column
ax1 = fig.add_subplot(gs[0:2,0])
ax1.plot(a)
ax1.set_aspect('equal')
#put a plot in the second row, first col
ax2 = fig.add_subplot(gs[2,2:3])
ax2.plot(b)
ax2.set_aspect('equal')
#have it second col, all rows
ax3 = fig.add_subplot(gs[4,4])
im = ax3.imshow(c)
fig.colorbar(im, ax=ax3, fraction=.045)
```

```python
#two different units on same graph

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(a)
for t in ax1.get_yticklabels():
    t.set_color('b')
    
ax2 = ax1.twinx()
ax2.plot(b, color='r')

for t in ax2.get_yticklabels():
    t.set_color('r')
#there's also a twiny
```
