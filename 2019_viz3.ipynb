```python
import numpy as np # we will use this later, so import it now

from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
```

```python
import pandas as pd
df = pd.read_csv("data/nyc_2005_2015.csv")
cp = df[df['STATION_NAME'].str.contains('NEW YORK NY US')][['DATE','MNTM']]
cp['DATE']
cp['date'] = pd.to_datetime(cp['DATE'], format="%Y%m%d")
cp['temp'] = (cp['MNTM']/10*9/5) + 32 
```

```python
# create a new plot (with a title) using figure
p = figure(plot_width=600, plot_height=400, 
           title="Weather in Central Park",
           x_axis_type='datetime')

# add a line renderer
p.line(cp['date'], cp['temp'], line_width=2)

p.yaxis.axis_label = 'Temperature (F)'
show(p) 
```
